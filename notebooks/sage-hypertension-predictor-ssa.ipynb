## Preprocessing 

```python
print("Ghana:", ghana.shape)
print("South Africa:", south_africa.shape)
```

```python
cols = [
    'id', 'q0001', 'q1009', 'q1010_mm', 'q1010_yyyy', 'q0104',
    'q1011',
    'q2501_s', 'q2501_d', 'q2502_s', 'q2502_d', 'q2503_s', 'q2503_d',
    'q2506', 'q2507', 'q2508', 'q2509',
    'q3001', 'q3002', 'q3007', 'q3008', 'q3011',
    'q3016', 'q3017', 'q3019', 'q3022', 'q3025', 'q3028',
    'q3012', 'q3013', 'pweight'
]

cols_sa = [c for c in cols if c != 'q1010_yyyy']

ghana_clean = ghana[cols].copy()
south_africa_clean = south_africa[cols_sa].copy()

print("Ghana:", ghana_clean.shape)
print("South Africa:", south_africa_clean.shape)
```

```python
import numpy as np

ghana_clean = ghana_clean.replace(' ', np.nan)
ghana_clean = ghana_clean.replace('-8', np.nan)

south_africa_clean = south_africa_clean.replace(' ', np.nan)
south_africa_clean = south_africa_clean.replace('-8', np.nan)

print("Done")
```

```python
bp_cols = ['q2501_s', 'q2502_s', 'q2503_s', 'q2501_d', 'q2502_d', 'q2503_d']

ghana_clean = ghana_clean.dropna(subset=bp_cols)
south_africa_clean = south_africa_clean.dropna(subset=bp_cols)

print("Ghana:", ghana_clean.shape)
print("South Africa:", south_africa_clean.shape)
```

```python
numeric_cols = [
    'q2501_s', 'q2501_d', 'q2502_s', 'q2502_d', 'q2503_s', 'q2503_d',
    'q2506', 'q2507', 'q2508', 'q2509',
    'q1010_mm', 'q1010_yyyy', 'q1011',
    'q3001', 'q3002', 'q3007', 'q3008', 'q3011',
    'q3016', 'q3017', 'q3019', 'q3022', 'q3025', 'q3028',
    'q3012', 'q3013', 'pweight'
]

for col in numeric_cols:
    if col in ghana_clean.columns:
        ghana_clean[col] = pd.to_numeric(ghana_clean[col], errors='coerce')
    if col in south_africa_clean.columns:
        south_africa_clean[col] = pd.to_numeric(south_africa_clean[col], errors='coerce')

print(ghana_clean.dtypes)
```

```python
skip_cols = ['q3002', 'q3008', 'q3011', 'q3017']

for col in skip_cols:
    ghana_clean[col] = ghana_clean[col].fillna(0)
    south_africa_clean[col] = south_africa_clean[col].fillna(0)

print("Done")
```

```python
small_null_cols = ['q3001', 'q3007', 'q3016', 'q3019', 
                   'q3022', 'q3025', 'q3028']

ghana_clean = ghana_clean.dropna(subset=small_null_cols)
south_africa_clean = south_africa_clean.dropna(subset=small_null_cols)

print("Ghana:", ghana_clean.shape)
print("South Africa:", south_africa_clean.shape)
```

```python
body_cols = ['q2506', 'q2507', 'q2508', 'q2509']

for col in body_cols:
    ghana_clean[col] = ghana_clean[col].fillna(ghana_clean[col].median())
    south_africa_clean[col] = south_africa_clean[col].fillna(south_africa_clean[col].median())

print("Done")
```

```python
# Diet columns - median
for col in ['q3012', 'q3013']:
    ghana_clean[col] = ghana_clean[col].fillna(ghana_clean[col].median())
    south_africa_clean[col] = south_africa_clean[col].fillna(south_africa_clean[col].median())

# Sex and residence - mode
south_africa_clean['q1009'] = south_africa_clean['q1009'].fillna(
    south_africa_clean['q1009'].mode()[0])
south_africa_clean['q0104'] = south_africa_clean['q0104'].fillna(
    south_africa_clean['q0104'].mode()[0])

# Survey weight - drop
ghana_clean = ghana_clean.dropna(subset=['pweight'])
south_africa_clean = south_africa_clean.dropna(subset=['pweight'])

print("Ghana:", ghana_clean.shape)
print("South Africa:", south_africa_clean.shape)
```

```python
rename_cols = {
    'id': 'id',
    'q0001': 'country',
    'q1009': 'sex',
    'q1010_mm': 'birth_month',
    'q1010_yyyy': 'birth_year',
    'q0104': 'residence',
    'q1011': 'age',
    'q2501_s': 'sbp1', 'q2501_d': 'dbp1',
    'q2502_s': 'sbp2', 'q2502_d': 'dbp2',
    'q2503_s': 'sbp3', 'q2503_d': 'dbp3',
    'q2506': 'height_cm', 'q2507': 'weight_kg',
    'q2508': 'waist_cm', 'q2509': 'hip_cm',
    'q3001': 'ever_tobacco', 'q3002': 'current_tobacco',
    'q3007': 'ever_alcohol', 'q3008': 'recent_alcohol',
    'q3011': 'avg_drinks_daily', 'q3016': 'vigorous_work',
    'q3017': 'days_vigorous_work', 'q3019': 'moderate_work',
    'q3022': 'walks_or_cycles', 'q3025': 'vigorous_leisure',
    'q3028': 'moderate_leisure', 'q3012': 'fruit_servings',
    'q3013': 'veg_servings', 'pweight': 'survey_weight'
}

ghana_clean = ghana_clean.rename(columns=rename_cols)
south_africa_clean = south_africa_clean.rename(columns=rename_cols)

print(ghana_clean.columns.tolist())
```

```python
south_africa_clean['birth_year'] = np.nan

df = pd.concat([ghana_clean, south_africa_clean], ignore_index=True)

print("Combined dataset:", df.shape)
print(df['country'].value_counts())
```

```python
print(df.isnull().sum())
```

```python
df.to_csv('../docs/df_preprocessed.csv', index=False)
print("Saved successfully")
```

## Preprocessing Summary

**Input:** Ghana (5,573) + South Africa (4,227) = 9,800 respondents
**Output:** Combined clean dataset of 8,777 respondents, 31 variables

### Steps Performed

1. Replaced blank spaces and -8 codes with NaN
2. Dropped rows missing blood pressure readings (6.2% of data)
3. Converted all string columns to numeric types
4. Filled skip logic nulls with 0 (current_tobacco, recent_alcohol,
   avg_drinks_daily, days_vigorous_work)
5. Imputed body measurements and diet variables with column median
6. Imputed sex and residence with column mode
7. Dropped rows with missing survey_weight
8. Added id and q1011 (direct age in years) to variable selection
9. Added birth_year as NaN to South Africa to maintain structure
10. Combined Ghana and South Africa into single dataframe

### Remaining Nulls
- birth_month and birth_year remain but are no longer needed
- Direct age variable (q1011) covers all 8,777 respondents

### Country Codes
- 241 = Ghana (4,960 respondents)
- 155 = South Africa (3,927 respondents)

### Key Decisions
- BP missing rows dropped — target variable cannot be imputed
- Skip logic nulls filled with 0 — absence of behaviour is a real value
- Median used over mean for body measurements — resistant to outliers
- Row count dropped from 9,800 to 8,777 — 10.4% total data loss,
  within acceptable limits for health research

## Day 6. Feature Engineering

```python
df = pd.read_csv('../docs/df_preprocessed.csv')
print(df.shape)
print(df.columns.tolist())
```

```python
df['sbp_avg'] = df[['sbp1', 'sbp2', 'sbp3']].mean(axis=1)
df['dbp_avg'] = df[['dbp1', 'dbp2', 'dbp3']].mean(axis=1)

# Fix -9 codes that slipped through
df['sbp_avg'] = df['sbp_avg'].replace(-9, np.nan)
df['dbp_avg'] = df['dbp_avg'].replace(-9, np.nan)
df = df.dropna(subset=['sbp_avg', 'dbp_avg'])

print(df['sbp_avg'].describe())
```

```python
def classify_bp(row):
    sbp = row['sbp_avg']
    dbp = row['dbp_avg']
    if sbp >= 180 or dbp >= 110:
        return 3  # Grade 3
    elif sbp >= 160 or dbp >= 100:
        return 2  # Grade 2
    elif sbp >= 140 or dbp >= 90:
        return 1  # Grade 1
    else:
        return 0  # Normal

df['hypertension_grade'] = df.apply(classify_bp, axis=1)
print(df['hypertension_grade'].value_counts())
```

```python
df['bmi'] = df['weight_kg'] / ((df['height_cm'] / 100) ** 2)

df = df[(df['bmi'] >= 10) & (df['bmi'] <= 70)]
df = df[(df['height_cm'] >= 100) & (df['height_cm'] <= 220)]
df = df[(df['weight_kg'] >= 30) & (df['weight_kg'] <= 200)]

print(df['bmi'].describe())
print("Rows remaining:", len(df))
```

```python
df['country_name'] = df['country'].map({241: 'Ghana', 155: 'South Africa'})
print(df['country_name'].value_counts())
```

```python
# Age groups
def age_group(age):
    if age < 50:
        return '18-49'
    elif age < 60:
        return '50-59'
    elif age < 70:
        return '60-69'
    elif age < 80:
        return '70-79'
    else:
        return '80+'

df['age_group'] = df['age'].apply(age_group)
print(df['age_group'].value_counts())
```

```python
# Encode sex: Male = 1, Female = 0
df['sex_encoded'] = df['sex'].map({1: 1, 2: 0})
print(df['sex_encoded'].value_counts())
```

```python
# Age-sex interaction
df['age_sex_interaction'] = df['age'] * df['sex_encoded']
print(df['age_sex_interaction'].describe())
```

```python
print(df['sex'].dtype)
print(df['sex'].unique())
```

```python
cols_to_drop = ['sbp1', 'sbp2', 'sbp3', 'dbp1', 'dbp2', 'dbp3', 
                'birth_month', 'birth_year']

df = df.drop(columns=cols_to_drop)
print(df.shape)
print(df.columns.tolist())
```

```python
df.to_csv('../docs/df_features.csv', index=False)
print("Saved successfully")
```
